# 🔧 Pump Failure Prediction — MVP

Predictive maintenance model for hydraulic piston pump valve plate failure
detection, developed for copper mining production environments.

## 🎯 Overview

This project implements a **4-class classification model** that diagnoses
the operational state of a hydraulic pump based on 7 sensor readings:

| Class | Description |
|---|---|
| ✅ Normal | Standard operation |
| ⚠️ Valve Plate Wear | Progressive degradation |
| 🔴 Simulated Failure 1 | First level forced failure |
| 🔴 Simulated Failure 2 | Second level forced failure |

**Final model performance (test set):**
- F1-Macro: **99.72%**
- ROC-AUC: **100.0%**
- False Negative Rate (Failure → Normal): **~0.0%**

## 🏗️ Architecture

```
Raw Data → Preprocessing → Feature Engineering → XGBoost → Dashboard
```

## 🚀 Quick Start

### 1. Clone the repository
```bash
git clone https://github.com/your-org/pump-failure-prediction.git
cd pump-failure-prediction
```

### 2. Install dependencies
```bash
pip install -r requirements.txt
```

### 3. Download the dataset
Download from [Kaggle](https://www.kaggle.com/datasets/mbjunior/valve-plate-failure-prediction-in-hydraulic-pumps)
and place CSV files in `data/raw/`.

### 4. Run the full pipeline
```python
# In a notebook or Python script
from src.data.loader import main as load_data
from src.data.preprocessor import run_preprocessing_pipeline
from src.features.engineer import run_feature_engineering
from src.models.trainer import run_training_pipeline
from src.models.evaluator import run_evaluation
from src.models.persistor import run_persistence_pipeline, save_model_metadata

load_data()
X_train, X_val, X_test, y_train, y_val, y_test = run_preprocessing_pipeline()
X_train, X_val, X_test = run_feature_engineering(X_train, X_val, X_test)
run_training_pipeline(X_train, y_train, X_val, y_val)
run_evaluation(X_test, y_test)
run_persistence_pipeline()
save_model_metadata(feature_names=X_train.columns.tolist())
```

### 5. Run the dashboard
```bash
streamlit run app/Home.py
```

### 6. Run MLflow UI
```bash
mlflow ui --backend-store-uri sqlite:///mlflow.db
```
